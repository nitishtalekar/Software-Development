```python
def clean(x):
#     if x == 'Ab':
#         return int(0)
    x = list(str(x))
    if x[-1] == 'E' or  x[-1] == 'F':
        return "".join(x[:-1]).strip()
    else:
        return "".join(x)
```

```python
for i in dfmarks.columns[1:-4]:
    dfmarks.loc[:,i] = dfmarks.loc[:,i].apply(clean)
```
