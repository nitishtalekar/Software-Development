```python
#LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#READ FILE + DROP
df = pd.read_excel('RA.xlsx', header = 7)
df = df.drop(['Unnamed: 1','Unnamed: 28','Unnamed: 29'], axis=1)
```

```python
df
```

```python
#HEADER LIST
col = list(df.columns.values)
c = col[:]
col2 = df.iloc[:1,:].values[0]

i=2
while i < len(col)-3:
    x = col[i-1]
    while 'Unnamed' in col[i]:
        col[i] = x
        i = i+1
    i = i+1
    
#INITIALIZE
col[0] = 'NAME'
col[i] = 'TOTAL'
col[i+1] = 'GPA'
col[i+2] = 'RESULT'
col2[0] = ''
col2[-1] = ''
col2[-2] = ''
col2[-3] = ''

# print(l)
cols = list(zip(col,col2))
header = [i[0]+'-'+i[1] for i in cols]
```

```python
df = df.rename(columns=dict(zip(c, header)))
df = df.iloc[3:,:]
df.index = pd.RangeIndex(len(df.index))
dfmarks = df.loc[::2]
dfnames = df.loc[1::2]


dfmarks['Name'] = dfnames['NAME-'].values

dfmarks.index = pd.RangeIndex(len(dfmarks.index))
```

```python
l = [i for i in dfnames.columns if 'TOT' in i][:-1] + list(dfnames.columns[-3:])
dfnames = dfnames.drop(l, axis=1)
```

```python
def clean(x):
    if x == 'Ab':
        return int(0)
    x = list(str(x))
    if x[-1] == 'E' or  x[-1] == 'F':
        return int("".join(x[:-1]))
    else:
        return int("".join(x))
```

```python
dfmarks.columns[1:-4]
```

```python
for i in dfmarks.columns[1:-4]:
    dfmarks[i] = dfmarks[i].apply(clean)
```
